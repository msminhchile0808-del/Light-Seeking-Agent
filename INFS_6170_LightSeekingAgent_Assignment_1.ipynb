```python
import random

class LightGrid:
    """
    Represents an environment with light sources.
    The grid where each cell can either contain a light source (True) or be dark (False).
    """

    def __init__(self, rows=5, cols=5):
        self.rows = rows
        self.cols = cols
        # Randomly mark each cell as having a light source (True) or being dark (False)
        self.grid = [
            [random.choice([True, False]) for _ in range(cols)]
            for _ in range(rows)
        ]

    def has_light(self, row, col):
        """Return True if the cell at (row, col) has a light source."""
        return self.grid[row][col]

    def absorb_light(self, row, col):
        """Mark the cell at (row, col) as dark (light absorbed)."""
        self.grid[row][col] = False

    def all_light_absorbed(self):
        """Return True if every cell in the grid is dark (goal achieved)."""
        return all(not cell for row in self.grid for cell in row)

    def display(self, agent_row, agent_col):
        """Print a visual snapshot of the grid with the agent's position."""
        print("\n  " + " ".join(str(c) for c in range(self.cols)))
        for r in range(self.rows):
            row_str = f"{r} "
            for c in range(self.cols):
                if r == agent_row and c == agent_col:
                    row_str += "A "   # A = Agent's current position
                elif self.grid[r][c]:
                    row_str += "* "   # * = Light source
                else:
                    row_str += ". "   # . = Dark cell
            print(row_str)
        print("  (* = light, . = dark, A = agent)\n")

class LightSeekingAgent:
    """
    A simple reflex agent that seeks and absorbs light sources in a LightGrid.

    Decision logic:
      - If the current cell has a light source → absorb it
      - If the current cell is dark → move to the next cell
    """

    def __init__(self, light_grid):
        self.light_grid = light_grid
        self.row = 0        # Starting row position (top-left corner)
        self.col = 0        # Starting column position (top-left corner)
        self.moves = 0      # Counter: how many times the agent moved
        self.lights_absorbed = 0 # Counter: how many light sources the agent absorbed

    def perceive(self):
        """
        Sense the environment at the current position.
        Returns True if the cell has a light source, False if it's dark.
        This is the agent's only input from the world.
        """
        return self.light_grid.has_light(self.row, self.col)

    def act(self):
        """
        Decide and perform an action based on the current perception.
        If light is present, absorb it; otherwise, move.
        """
        if self.perceive():
            # The cell has a light source: perform the ABSORB action
            self.light_grid.absorb_light(self.row, self.col)
            self.lights_absorbed += 1
            print(f" qAbsorbed light at ({self.row},{self.col})")
        else:
            # The cell is dark: perform the MOVE action
            self._move()

    def _move(self):
        """
        Move to the next cell using a left-to-right, top-to-bottom sweep.
        This ensures every cell gets visited exactly once.
        """
        if self.col + 1 < self.light_grid.cols:
            # Move right within the same row
            self.col += 1
        elif self.row + 1 < self.light_grid.rows:
            # Reached the end of a row: wrap to the start of the next row
            self.row += 1
            self.col = 0
        # If at the last cell (bottom-right), stay put — the job is done
        self.moves += 1

    def run(self):
        """
        Main loop: repeatedly perceive and act until all light is absorbed
        or the step limit is reached (safety guard against infinite loops).
        """
        print("=== Light-Seeking Agent Starting ===")
        self.light_grid.display(self.row, self.col)

        # Set a maximum number of steps to prevent infinite loops
        steps = self.light_grid.rows * self.light_grid.cols * 2
        for step in range(steps):
            print(f"Step {step + 1} — Position: ({self.row},{self.col})")
            self.act()                              # Perceive → Decide → Act
            self.light_grid.display(self.row, self.col)  # Show updated grid state

            # Check if the agent has achieved its goal
            if self.light_grid.all_light_absorbed():
                print(" All light absorbed!")
                break
        else:
            # This runs only if the loop wasn't broken by all_light_absorbed()
            print("Reached step limit.")

        # Print a summary of the agent's performance
        print(f"\n Summary: {self.lights_absorbed} light sources absorbed in {self.moves} moves.")


if __name__ == "__main__":
    random.seed(42)                  # Fixed seed for reproducible results
    light_grid = LightGrid(rows=4, cols=4)      # Create a 4x4 room with random light sources
    agent = LightSeekingAgent(light_grid)        # Create the light-seeking agent
    agent.run()                      # Start the agent
```
